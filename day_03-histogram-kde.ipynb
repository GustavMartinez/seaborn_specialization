### Histogramas y KDEs

```python
# Import libraries

import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Cargar datos
tips = sns.load_dataset("tips")
```

### Personalización de colores y estilos

```python
# change style: sns.set_style()
# Este comado permite cambiar el estilo general de los gráficos.
sns.set_style("darkgrid")

# other set_styles: "whitegrid", "darkgrid", "white", "dark", "ticks".


# Change palette color: sns.set_palette()
# Este comando se utiliza para establecer la paleta de colores de los gráficos.
sns.set_palette("muted")

# Other set_palette: "deep", "muted", "pastel", "bright", "dark", "colorblind"
```

```python
sns.scatterplot(x="total_bill", y="tip", data=tips)
plt.show()
```

#### Histogramas
Los histogramas son una buena herramienta para visualizar la distribución de una variable continua.<br>
Un histograma divide los datos en intervalos (o bins) y muestra la frecuencia de puntos de datos dentro de cada intervalo.<br>
La altura de cada barra representa la cantidad de datos en ese rango.


```python
# Histograma básico

sns.histplot(data=tips, x="total_bill")
plt.title("Histograma básico del total de la cuenta")
plt.show()
```

```python
# Controlar el número de bins del histograma: bins=

sns.histplot(data=tips, x="total_bill", bins=25)
plt.title("Histograma básico del total de la cuenta")
plt.show()
```

```python
# En lugar de controlar el número de barras, se puede definir el ancho de cada barra con "binwidth"

sns.histplot(data=tips, x="total_bill", binwidth=10)
plt.show()
```

```python
# Junto con el binwidth, se puede utilizar el binrange(), para definir inicio e intervalo

sns.histplot(
        data=tips,
        x='total_bill',
        binwidth=5,
        binrange=(0,55)
)

plt.show()
```

```python
# Histograma combinado con KDE

sns.histplot(
        data=tips, 
        x='total_bill',
        binwidth=5,
        binrange=(0,50), 
        kde=True
)

plt.title("Histograma con KDE")
plt.show()
```

```python
# Colores y transparencia del histograma

sns.histplot(
        data=tips, 
        x="total_bill", 
        color='green',
        binwidth=5,
        binrange=(0,50), 
        alpha=0.3
)

plt.title("Histograma personalizado")
plt.show()
```

```python
# Elementos visuales: "bars" (default), "step", "poly"

sns.histplot(
        data=tips, 
        x="total_bill", 
        color= 'blue',
        alpha=0.2,
        binwidth= 2.5,
        binrange=(5,50),
        element="step"
)

plt.title("Histograma con estilo 'Step'")
plt.show()
```

```python
# Elementos visuales: "bars" (default), "step", "poly"

sns.histplot(
    data=tips, 
    x='total_bill',
    color='green', 
    alpha=0.3,
    binwidth=5,
    binrange=(5,50),
    element='poly')

plt.title("Histograma con estilo 'poly'")
plt.show()
```

```python
# Separación por categorías (hue)
# El argumento hue, permite dividir los datos por categorias y mostrar diferentes colores para cada grupo.

sns.histplot(
        data=tips,
        x='total_bill',
        hue='sex',
        binwidth=2.5,
        binrange=(5,50),
        multiple="stack"
)

# Opciones para multiple: layer (Superpone las categorias); stack (Apila las barras); fill (Normaliza las alturas para que sumen 1)

plt.title("Histograma separado por sexo. Stack")
plt.show()
```

```python
# Separación por categorías (hue)
# El argumento hue, permite dividir los datos por categorias y mostrar diferentes colores para cada grupo.

sns.histplot(
        data=tips,
        x='total_bill',
        hue='sex',
        multiple='layer'
)

# Opciones para multiple: layer (Superpone las categorias); stack (Apila las barras); fill (Normaliza las alturas para que sumen 1)

plt.title("Histograma separado por sexo. Layer")
plt.show()
```

```python
# Separación por categorías (hue)
# El argumento hue, permite dividir los datos por categorias y mostrar diferentes colores para cada grupo.

sns.histplot(
        data=tips,
        x='total_bill',
        bins=20,
        hue='sex',
        multiple='fill'
)

# Opciones para multiple: layer (Superpone las categorias); stack (Apila las barras); fill (Normaliza las alturas para que sumen 1)

plt.title("Histograma separado por sexo. fill")
plt.show()
```

```python
# El borde y grosor de las barras se puede ajustar con 'linewidth' y 'edgecolor'

sns.histplot(
        data=tips,
        x='total_bill',
        hue='sex',
        multiple='stack',
        binwidth=2.5,
        binrange=(5,50),
        edgecolor='black',
        linewidth=0.5,
        alpha=0.5
)



plt.title("Histograma con bordes personalizados")
plt.show()
```

```python
# Escala del eje Y
# El parámetro 'stat' controla cómo se escala el eje Y:

sns.histplot(
        data=tips,
        x='total_bill',
        stat='density',
        alpha=0.5,
        linewidth=0.5,
        edgecolor='black',
)


# Opciones para stat: count (default - muestra la frecuencia). density (Normaliza para que el área total sea 1). probability (Normaliza para que las alturas sumen 1)

plt.title("Histograma escalado por densidad")
plt.show()
```

```python
# Escala del eje Y
# El parámetro 'stat' controla cómo se escala el eje Y:

sns.histplot(
        data=tips,
        x='total_bill',
        stat='probability',
        binwidth=2.5,
        binrange=(5,50),
        linewidth=0.5,
        edgecolor='black',
        color='purple',
        alpha=0.4
)

# Opciones para stat: count (default - muestra la frecuencia). density (Normaliza para que el área total sea 1). probability (Normaliza para que las alturas sumen 1)

plt.title("Histograma escalado por probabilidad")
plt.show()
```

### Gráfico de densidad (KDE)

El gráfico de densidad Kernel Density Estimation (KDE) es una herramienta visual para estimar la distribución de una variable continua.<br>
En lugar de mostrar las frecuencias como un histograma, el KDE representa la densidad de probabilidad de los datos como una curva suavizada.<br>
Es ideal para observar patrones generales en los datos sin la segmentación en intervalos que tienen los histogramas.

```python
# Gráfico de densidad (KDE) básico

sns.kdeplot(
        data=tips, 
        x="total_bill"
)

plt.title("KDE Plot")
plt.xlabel("Total Bill")
plt.show()
```

```python
# Adicionar sombreado bajo la curva con fill

sns.kdeplot(
        data=tips,
        x='total_bill',
        fill=True
)

# fill = true

plt.title("Gráfico KDE con sombreado")
plt.show()
```

```python
# Ancho del núcleo:
# El ancho del nucleo bandwidth controla el grado de suavizado. un valor más pequeño produce una curva más ajustada a los datos;
# un valor más grande genera una curva más suave.

sns.kdeplot(
    data=tips,
    x='total_bill',
    bw_adjust=0.5, # más ajustado
    color='red'
)

sns.kdeplot(
    data=tips,
    x='total_bill',
    bw_adjust=2 # más suave
)

plt.title("Ajuste del ancho al núcleo")
plt.show()
```

```python
sns.kdeplot(
        data=tips, 
        x="total_bill", 
        hue="sex", 
        fill=True)
plt.title("Distribución KDE por sexo")
plt.show()
```

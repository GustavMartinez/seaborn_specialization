# Gráficos básicos en Seaborn - Day 2

```python
# Import libraries:

import seaborn as sns
import matplotlib.pyplot as plt

# cargar datos de ejemplo
titanic = sns.load_dataset('titanic')
```

### Gráficos de barras
Los gráficos de barras son útiles para comparar cantidaddes entre diferentes categorias.

```python
# Crear un gráfico de barras
sns.barplot(x = 'class', y='fare', data=titanic)

# Mostrar el gráfico
plt.show()
```

### Gráfico de caja (Boxplots)
Los boxplots muestran la distribución de datos numéricos y son útiles para detectar valores atípicos.

```python
# Crear un boxplot

sns.boxplot(x = "class", y="age", data=titanic)

# Mostrar el gráfico
plt.show()
```

### Gráfico de violín

Los gráficos de violín son similares a los boxplots, pero muestran la densidad de probabilidad de los datos.

```python
# Crear un gráfico de violín

sns.violinplot(x="class", y="age", data=titanic)

# Mostrar el gráfico
plt.show()
```

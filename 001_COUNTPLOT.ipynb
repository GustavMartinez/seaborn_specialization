```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('/home/gustavo/Documents/Study/007_skills/003_Seaborn/data/BankChurners.csv')
```

```python
data.head()
```

## Task: Create a basic countplot for each of the 'Card Category'

```python
plt.figure(figsize=(13,8))

sns.countplot(data=data, x='Card_Category')

plt.show()
```

## Options

### Color and style

Palette: change color palette <br>
Style: sns.set_style('whitegrid'); sns,set_theme(style='ticks')

```python
plt.figure(figsize=(13,8))

sns.set_style('whitegrid')

sns.countplot(data=data, x='Card_Category', hue='Card_Category', palette='pastel')


plt.show()
```

### Rotate ticks

```python
plt.figure(figsize=(13,8))

sns.countplot(data=data, x= "Card_Category", hue="Card_Category", palette='viridis')

plt.xticks(rotation=45)

plt.show()
```

### Bar order

```python
plt.figure(figsize=(13,8))

sns.countplot(data=data, x="Card_Category", hue='Card_Category', palette='inferno', order=['Blue', 'Silver', 'Gold', 'Platinum'])

plt.xticks(rotation=45)

plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

# Data

## Seaborn built-in datasets:

```python
print(sns.get_dataset_names())
```

## Import car craches:

```python
crash_df = sns.load_dataset('car_crashes')
crash_df.head()
```

## Distribution plot
Univariate distributions

```python
sns.displot(crash_df['not_distracted'], kde=False, bins=15)
```

## Join Plot

```python
sns.jointplot(x='speeding', y='alcohol', data=crash_df, kind='reg')
```

```python
sns.jointplot(x='speeding', y='alcohol', data=crash_df, kind='kde')
```

```python
sns.jointplot(x='speeding', y='alcohol', data=crash_df, kind='hex')
```

## KDE plot

```python
sns.kdeplot(crash_df['alcohol'])
```

## Pair Plots

```python
sns.pairplot(crash_df, hue='abbrev', palette='Blues')
```

## Rug plot

```python
sns.rugplot(crash_df['speeding'])
```

## Styling

```python
sns.set_style('ticks')

"""
Set styles:

white
whitegrid
dark
darkgrid
ticks

"""

plt.figure(figsize=(8,4))

sns.set_context('paper', font_scale=1.4)

sns.jointplot(x='speeding', y='alcohol', data=crash_df, kind='reg')
```

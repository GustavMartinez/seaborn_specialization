### Countplot

El countplot es un gráfico de barras que muestra el conteo de observaciones en cada categoria de una variable categorica.<br>
Distribución de categorias: Ideal para ver cómo se distribuyen las observaciones entre diferentes categorias.<br>
Análisis de encuestas: Útil para visualizar respuestas a preguntas de opción múltiple.<br>


```python
# Import libraries
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Load dataset
titanic = sns.load_dataset("titanic")
```

```python
titanic.head()
```

```python
# Set style and palette

sns.set_style('darkgrid')

# Set palette
sns.set_palette('pastel')
```

### Basic countplot

```python
plt.figure(figsize=(8,6)) # tamaño de la figura

sns.countplot(
        data=titanic,
        x='class',
        hue='class'
)

plt.xticks(rotation=25)

plt.title("CountPlot of class")

plt.show()
```

```python
# Define order of the classes:

plt.figure(figsize=(8,6))

sns.countplot(
        data=titanic,
        x='class',
        hue='class',
        order=["Third", "Second", "First"]
)


plt.xticks(rotation=20)

plt.title("Countplot with specific order")

plt.show()
```

```python
# Seaborn ofrece varias paletas de colores predefinidas: palette=

plt.figure(figsize=(8,6))

sns.countplot(
        data=titanic,
        x='class',
        hue='class',
        order=['Third', 'Second', 'First'],
        palette='dark:skyblue'
)

plt.xticks(rotation=25)

plt.title("Countplot with specific order and colors")
plt.xlabel("Class")
plt.ylabel("Count")

plt.show()
```

```python
# El gráfico puede ser en sentido vertical (default) en x. O en sentido horizontal en y.

plt.figure(figsize=(8,6))

sns.countplot(
        data=titanic,
        y='class', # Define y for the categories
        hue='class',
        palette='dark:red',
        order=['Third', 'Second', 'First'],
        alpha=0.7
)

plt.yticks(rotation=25)

plt.xlabel("Count")
plt.ylabel("Class")
plt.title("Countplot in horizontal")

plt.show()
```

```python
# Añadir una segunda variabel categórica con 'hue'

plt.figure(figsize=(8,6))

sns.countplot(
        data=titanic,
        x='class', # Variable principal
        hue='sex', # Segunda variable
        order=['Third', 'Second', 'First'],
        palette='dark:green',
        alpha=0.6
)

plt.xticks(rotation=25)

plt.title("Countplot by class and a second variable")
plt.xlabel("Class")
plt.ylabel("Count")

plt.show()
```

```python
# Ajustar el tamaño de las barras

plt.figure(figsize=(10,8))

sns.countplot(
        data=titanic,
        x='class',
        hue='sex',
        palette='dark:yellow',
        alpha=0.5,
        order=["Third", "Second", "First"],
        width=0.25
)

plt.xticks(rotation=25)

plt.title("Countplot with specific bar width")
plt.xlabel("Class")
plt.ylabel("Count")

plt.show()
```

```python
# Añadir etiquetas de datos

plt.figure(figsize=(8,6))

ax = sns.countplot(
        data=titanic,
        x='class',
        hue='sex',
        palette='dark:skyblue',
        alpha=0.4,
        order=["Third", "Second", "First"],
        width=0.3,
        linewidth=0.7, # tamaño del borde de las barras
        edgecolor='black' # Color del borde de las barras
)

for container in ax.containers:
    ax.bar_label(container)

plt.xticks(rotation=25)

plt.title("Countplot with labels on bars")
plt.xlabel("Class")
plt.ylabel("Count")

plt.show()
```
